```python
import nltk
a =  "i am very happy but not solved"
nltk.download('punkt')
```

```python
from nltk.tokenize import word_tokenize
token = word_tokenize(a)
token
```

```python
import spacy
nlp = spacy.blank('en')
stop = nlp.Defaults.stop_words
len(stop)
```

```python
l = []
for word in token:
  if word not in stop:
    l.append(word)
l
```
