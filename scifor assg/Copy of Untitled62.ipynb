```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.DataFrame()
df['year'] = np.arange(2018,2023,1)
df['place'] = np.array(['chennai','Hyderabad','punjab','Hyderabad','chennai'])
df['score'] = np.array([206,305,281,432,321])
df
```

```python
label = LabelEncoder()
df['place'] = label.fit_transform(df['place'])
df
```

```python
x = df[['year','place']]
y = df[['score']]
```

```python
#Linear Regression
model1 = LinearRegression()
model1.fit(x,y)
```

```python
# i simply done not testing and going for prediction
# Assume final match is on Chennai for the year 2023
model1.predict([[2023,1]])
```

```python
#LogisticRegression
#Cash price more than five lakhs
df['Cash price more than five lakhs'] = np.array(['yes','no','no','yes','yes'])
df
```

```python
df['Cash price more than five lakhs'] = label.fit_transform(df['Cash price more than five lakhs'])
df
```

```python
x1 = df[['year','place','score']]
y1 = df[['Cash price more than five lakhs']]
```

```python
model2 = LogisticRegression()
model2.fit(x1,y1)
```

```python
model2.predict([[2023,1,405.6]])
# It means yes,he will win cash price more than five lakhs
```

```python
#Decisiontreeregressor
model3 = DecisionTreeRegressor()
model3.fit(x,y)
```

```python
# i simply done not testing and going for prediction
# Assume final match is on Chennai for the year 2023
model3.predict([[2023,1]])
# See the changes,the result is last record result because attributes and records are less (no sufficient data)
```

```python
#Support vector Regression
model4 = SVR()
model4.fit(x,y)
```

```python
# i simply done not testing and going for prediction
# Assume final match is on Chennai for the year 2023
model4.predict([[2023,1]])
# Note it the result is changing from LinearRegression
```
