```python
x = ['Snack','milk','fruit','vegetables','flour']
y = [180,230,450,320,630]
plt.bar(x,y,color=['red','green'])
plt.title('ABC Company in year 2022')
plt.xlabel('category')
plt.ylabel('production count')
```

```python
x = ['Snack','milk','fruit','vegetables','flour']
y = [180,230,450,320,630]
plt.barh(x,y,color=['red','green'])
plt.title('ABC Company in year 2022')
plt.xlabel('category')
plt.ylabel('production count')
```

```python
x = ['Snack','milk','fruit','vegetables','flour']
y = [180,230,450,320,630]
plt.stackplot(x,y, labels=['uu',color=['green'])
plt.title('ABC Company in year 2022')
plt.xlabel('category')
plt.ylabel('production count')
plt.legend()
```
