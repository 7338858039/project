```python
import gym
from gym import spaces,Env
import numpy as np
```

```python
class navin(Env):
  def __init__(self):
    self.observation_space = [(i,j) for i in range(4) for j in range(4)]
    self.action_space = spaces.Discrete(4)
    self.state = self.observation_space[0]
    self.reward = 0
    # [left,right,top,bottom]

  def step(self,action):
    while True:
      if self.state == self.observation_space[8] or self.state == self.observation_space[12] or self.state == self.observation_space[5] or self.state == self.observation_space[1]:
        break
      else:
        self.reward+=1
    return self.reward
  def reset(self):
    return self.state
```

```python
modal = navin()
```

```python
value = modal.action_space.sample()
modal.step(value)
```
